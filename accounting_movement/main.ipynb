```python
import pandas as pd
from pathlib import Path
from datetime import date
```

Выгрузка из СБИС находится в папке Входящие. Необходимо загрузить все csv файлы в один датафрейм. Если встретился файл с другим расширением, то его пропускаем

```python
p=Path('.')/'Входящие'
csv_files=[file for file in p.rglob('*.csv')]

dflist=[]
for path in csv_files:
    try:
        tab=pd.read_csv(path, encoding='1251', sep=';')
        dflist.append(tab)
    except:
        print(f'problem in {path}')
    
sbis=pd.concat(dflist, ignore_index=True)

sbis
```

Дать столбцам полученного df такие названия.  
Все названия из нескольких слов сцепить через нижнее подчеркивание в одно_длинное_слово.

```python
needed_headers_1=[
    "Дата",
    "Номер",
    "Сумма",
    "Статус",
    "Примечание",
    "Комментарий",
    "Контрагент",
    "ИНН/КПП",
    "Организация",
    "ИНН/КПП",
    "Тип документа",
    "Имя файла",
    "Дата",
    "Номер 1",
    "Сумма 1",
    "Сумма НДС",
    "Ответственный",
    "Подразделение",
    "Код",
    "Дата",
    "Время",
    "Тип пакета",
    "Идентификатор пакета",
    "Запущено в обработку",
    "Получено контрагентом",
    "Завершено",
    "Увеличение суммы",
    "НДC",
    "Уменьшение суммы",
    "НДС"
]

needed_headers=[name.replace(' ','_') for name in needed_headers_1]

sbis.columns=needed_headers
sbis
```

Файлы с выгрузкой из аптек лежат в папке Аптеки/csv/correct/.

Загрузить на обработку по одному файлу. НЕ csv формат - игнорировать.

1. Если "Поставщик" - ЕАПТЕКА, то к "Номер накладной" нужно добавить '/15'.
2. Нужно найти все записи в выгрузке из СБИСа по данному номеру накладной
3. Из найденных строк нужно оставить только те, которые имеют один из типов документа: ["СчФктр", "УпдДоп", "УпдСчфДоп", "ЭДОНакл"]
4. Если ничего не найдено - просто переходим к следующей строке
5. Если найдено - нужно сохранить значения Номер, Сумма и Дата
6. Дату нужно представить в формате 25.05.2021
7. В столбцы из пункта 5 нужно записать найденные для данной строки значения
8. В столбец Сравнение дат помещаем "Не совпадает!", если найденная дата и дата накладной отличаются. Иначе - пустая строка.

Внимание:  
если по заданным условиям находится несколько подходящих документов - выбираем значения из первого найденного. Это означает, что если у документов разные суммы, берется та, что стоит первой в выгрузке из СБИС.

Итоговый файл выгрузки по аптеке должен содержать такие столбцы в заданном порядке:

'№ п/п', 'Штрих-код партии', 'Наименование товара', 'Поставщик',
'Дата приходного документа', 'Номер приходного документа',
'Дата накладной', 'Номер накладной', 'Номер счет-фактуры',
'Сумма счет-фактуры', 'Кол-во',
'Сумма в закупочных ценах без НДС', 'Ставка НДС поставщика',
'Сумма НДС', 'Сумма в закупочных ценах с НДС', 'Дата счет-фактуры', 'Сравнение дат'

Сохранить по пути Результат/{полная сегодняшняя дата}/{имя исходного файла без расширения} - результат.xlsx. Если таких папок не существует - создать автоматом.

```python
# выгружаем исходники по очереди
p2=Path('.') / 'Аптеки'/'csv'/'correct'
for path in p2.rglob('*.csv'):
    apt=pd.read_csv(path, sep=';', encoding='1251')

    # создаем пустые столбцы 
    new_cols=[
    "Номер счет-фактуры",
    "Сумма счет-фактуры",
    "Дата счет-фактуры",
    "Сравнение дат"
    ]
    apt[new_cols]=''

    # добавляем суффикс в номер накладной
    apt.loc[apt['Поставщик'].str.contains('ЕАПТЕКА', na=False), 'Номер накладной']+='/15'
    
    # выбираем в папке СБИС только записи с определенным типом документов 
    doc_types=["СчФктр", "УпдДоп", "УпдСчфДоп", "ЭДОНакл"]
    sbis_mask=sbis['Тип_документа'].isin(doc_types)
    sbis_filtered=sbis[sbis_mask].groupby(['Номер'], as_index=False).first().set_index('Номер')
    
    # найдем номер первого столбца с названием ДАТА для маппинга (следующий шаг)
    date_num=list(sbis.columns).index('Дата')
    
    # подтянем данные из СБИС в df АПТЕКА и проведем проверку
    apt['Номер счет-фактуры']=apt['Номер накладной'].map(sbis_filtered.index.to_series())
    apt['Сумма счет-фактуры']=apt['Номер накладной'].map(sbis_filtered['Сумма'])
    apt['Дата счет-фактуры']=pd.to_datetime(apt['Номер накладной'].map(sbis_filtered.iloc[:,date_num]), format='%d.%m.%y').dt.strftime('%d.%m.%Y')
    apt.loc[apt['Дата накладной']!=apt['Дата счет-фактуры'], 'Сравнение дат']="Не совпадает!"
    
    # правильно расположим столбцы, согласно ТЗ
    tab_to_safe=apt[['№ п/п', 'Штрих-код партии', 'Наименование товара', 'Поставщик',
    'Дата приходного документа', 'Номер приходного документа',
    'Дата накладной', 'Номер накладной', 'Номер счет-фактуры',
    'Сумма счет-фактуры', 'Кол-во',
    'Сумма в закупочных ценах без НДС', 'Ставка НДС поставщика',
    'Сумма НДС', 'Сумма в закупочных ценах с НДС', 'Дата счет-фактуры', 'Сравнение дат'
    ]]
    
    # создадим папку для результата, если нет
    safe_folder=Path('.')/'Результат'/f"{date.today()}"
    safe_folder.mkdir(parents=True, exist_ok=True)
    
    # сохраним результат в файл
    tab_to_safe.to_excel(safe_folder/f"{path.stem} - результат.xlsx", index=False)
    
    # проверка shape (отладка)
    print(tab_to_safe.shape)
```
